```python
import json
```

```python
test = open('ntnu.json',encoding="utf8")
data = json.load(test) #讀入資料
```

```python
type(data) #確認資料類型
```

```python
data.items()
```

```python
dep = data.get("師大")
dep
```

```python
ask = '地球科學系' #查詢科系學生總計
```

```python
for i in dep:
    if ask == i.get('科系'):
        print(ask,':', i.get('學生總計'), '個學生')
```

```python
all = []
for i in dep:
    all.append(i.get('學生總計'))  #排序學生總計
```

```python
all_enumerate = enumerate(all, start = 1)
all_enumerate_sorted = sorted(all_enumerate, key = lambda s: s[1], reverse = True)
```

```python
for i in all_enumerate_sorted:
    j = i[0] - 1
    print(dep[j].get('科系'),':', dep[j].get('學生總計'), '個學生') #印出排序後學生總計
```

```python
allw = []
for i in dep:
    allw.append(i.get('女生計')) #排序女生總計
```

```python
allw_enumerate = enumerate(allw, start = 1)
allw_enumerate_sorted = sorted(allw_enumerate, key = lambda s: s[1], reverse = True)
```
