```python
import jieba
import jieba.analyse
import pandas as pd
from collections import Counter
data = pd.read_csv("ba.csv")
data
```

```python
data['seg'] = ""
data.head()
```

```python
jieba.analyse.set_stop_words('stopwords.txt') #stopwords 自己不要的字的字典(自己建立)
index = 0
for documents in data['title']: #從 csv. 檔中的 JOB_DETAIL 欄位中找尋資料
    seg_list = jieba.analyse.extract_tags(documents, 150) #150 為前150個權重最高的單詞
    data['seg'].iloc[index] = seg_list
    index += 1
```
