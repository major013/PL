```python
# 比較台大跟政大的系所(學士) 

import pandas as pd #讀csv
NTU = pd.read_csv("NTU.csv")
NCCU =pd.read_csv("NCCU.csv")
```

```python
NTU.head()
```

```python
NCCU.head()
```

```python
NTU_D = list(NTU['科系名稱'])
print('台大的科系有:',len(NTU),'個')
NCCU_D = list(NCCU['科系名稱'])
print('政大的科系有:',len(NCCU),'個')
```

```python
print(NTU_D)
print(NCCU_D)
```

```python
#將list轉換成set

d1 = set(NTU_D)
d2 = set(NCCU_D)
```

```python
# 交集  intersection (台大跟政大都有的科系)

print('台大跟政大都有的科系')
print(d1&d2,'共',len(d1&d2),'個')
```

```python
# 聯集  union (所有台大跟政大的科系)

print('所有台大跟政大的科系')
print(d1|d2,'共',len(d1|d2),'個')
```

```python
# 差集  difference (台大有但政大沒有的科系)

print('台大有但政大沒有的科系')
print(d1-d2,'共',len(d1-d2),'個')
```

```python
# 差集  difference (政大有但台大沒有的科系)

print('政大有但台大沒有的科系')
print(d2-d1,'共',len(d2-d1),'個')
```

```python
# 對稱差集  symmetric_difference (只有台大有+只有政大有的科系)

print('只有台大有+只有政大有的科系')
print(d1^d2,'共',len(d1^d2),'個')
```
